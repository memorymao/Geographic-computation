#### **Task 1 Read in a data file of all counties in the US.**
###### Make a list of unique county names.

```python
def LoadJson():
     import requests, json
     url = "https://eric.clst.org/assets/wiki/uploads/Stuff/gz_2010_us_050_00_5m.json"

     # get the json data from the URL
     with requests.get(url) as response:
           data = response.json()
     return data

def JsonfileToData():
     data = LoadJson()
     features = data['features']
     # 1. Use + to concatenate strings
     print('Number of Counties in US: ' + str(len(features)))
     # 2. Use format function
     print('The first entry is {}'.format(features[0]['properties']['NAME']))
     # 3. use f to print
     print(f'The second entry is {features[1]["properties"]["NAME"]}')
     return features


# The following part is not used to implement the tasks in the assignment
# The following part is not essential for further tasks in the assignment, just added to better understand the structure of the input data
def dataStructureAnalysis(cols, data):
    import pandas as pd
    df =  pd.DataFrame(columns = cols)
    for f in data:
          #print(f['properties']['NAME'])
          df0 = pd.DataFrame(f['properties'], index=[0])
          df = pd.concat([df[:], df0[:]])
    return df

# main procedure:
data = JsonfileToData()
cols = ['GEO_ID','STATE', 'COUNTY', 'NAME', 'LSAD', 'CENSUSAREA']
dataStructureAnalysis(cols, data)
```

#### **Task 2 Find the three most common names of the counties.**
Derive the numbers of counties that use these three names, respectively. For each of them, list their county name and state code.

Hint: Think about what data structure would be best for this task. Once again, do not use special functions or packages for this task. Just use basic Python data structures and loops.

```python
def mostCommonName(k):
  featureList = JsonfileToData()
  cntyDict = {}

  for f in featureList:                             # f will be an individual feature

      name = f['properties']['NAME']               # name will be the county name

      if cntyDict.get(name) is None:               # find the name for the first time, mark it as 1
          cntyDict.update({name: 1})
      else:
          n = cntyDict.get(name)
          cntyDict.update({name: n+1})

# Once you have the dictionary cntyDict. Sort it by value.
  cntyDictSorted = sorted(cntyDict.items(), key=lambda item:item[1], reverse=True)
  print('The most common names of the counties are {}'.format([x[0] for x in cntyDictSorted[0:k]]))


# main procedure: get the three most common names of countries
mostCommonName(3)
```

#### **Task 3 Basic statistics by state.**
For each state, find

The number of counties
The name and size (census area) of the biggest and smallest county by area
The total and average area of counties

```python
def AreaDetailInfo():  # filter the area info for each state
    featureList = JsonfileToData()
    cntyDict = {}

    for f in featureList:                             # f will be an individual feature

        STATE = f['properties']['STATE']

        if cntyDict.get(STATE) is None:               # find the name for the first time, mark it as 1
            cntyDict[STATE]= []

        cntyDict[STATE].append([f['properties']['NAME'],f['properties']['CENSUSAREA']])

    return cntyDict


def AreaEachState(arr):    # find max/min/total/average value for each state
    total = sum([item[1] for item in arr])
    average = total/len(arr)
    cntyDictSorted = sorted(arr, key=lambda item:item[1], reverse=True)
    return cntyDictSorted[0], cntyDictSorted[-1], total, average


def AreaStatistic(cntyDict):
    maxstat = {}
    minstat = {}
    totalstat = {}
    avestat = {}
    for c in cntyDict:
      maxstat[c], minstat[c], totalstat[c], avestat[c]= AreaEachState(cntyDict[c])

    return maxstat, minstat, totalstat, avestat

def print_table(data):
    for row in data:
      #print('%s   {0:<8s}  $%.2f' % (row,data[row][0],data[row][1]))
      print(row, "   ",'{0:<24s}'.format(data[row][0]),data[row][1])



# main procedure:
cntyDict = AreaDetailInfo()
print(f'The total number of states is {len(cntyDict)}')
maxstat, minstat, totalstat, avestat = AreaStatistic(cntyDict)
print("The name and size (census area) of the biggest county by area are:")
print_table(maxstat)
print("The name and size (census area) of the smallest county by area are:")
print_table(minstat)
print("The total area of counties are:")
for row in totalstat:
  print('%s   $%.2f' % (row,totalstat[row]))
print("The average area of counties are:")
for row in avestat:
  print('%s   $%.2f' % (row,avestat[row]))
```
